```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 1. Load the dataset
# Replace 'results.csv' with your actual file path
df = pd.read_csv('random_cooked.csv')
```

```python
# 2. Map the binary variables to readable strategy names
def map_strategy(row):
    r, c = row['RAND'], row['COOCKED']
    if r == 0 and c == 0: return "Deterministic Modularity"
    if r == 1 and c == 0: return "Random Merge"
    if r == 0 and c == 1: return "All in One"
    return "Random + Cooked\n(RAND=1, COOCKED=1)"
    
```

```python
df['Strategy'] = df.apply(map_strategy, axis=1)
```

```python
# 3. Define the metric to plot
# Changing this to 'max_degree' if that column exists in your actual file
metric_to_plot = 'max_depth' 
```

```python
# strategy_order = ["Deterministic", "Weighted Random", "Cooked (Bad)", "Random + Cooked"]
```

```python

# 4. Create the plot
plt.figure(figsize=(10, 6))

# Create a bar plot with standard deviation error bars (useful since you have multiple SEEDs)
ax = sns.barplot(
    x='Strategy', 
    y=metric_to_plot, 
    data=df, 
    palette='Set2', 
    errorbar='sd',  # Displays standard deviation as error bars
    capsize=0.1
)

# Formatting the chart
plt.title('Max Depth per Merge Strategy', fontsize=14, pad=15)
plt.ylabel('Max Depth', fontsize=12)
plt.xlabel('Merge Strategy', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Add exact value labels on top of the bars
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        ax.annotate(f'{height:.1f}', 
                    (p.get_x() + p.get_width() / 2., height), 
                    ha='center', va='bottom', 
                    fontsize=10, color='black', xytext=(0, 5), 
                    textcoords='offset points')

plt.tight_layout()
plt.show()
```
